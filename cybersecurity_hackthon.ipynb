```python
import os

class WhisperingGuardian:
    def __init__(self, scan_path):
        self.scan_path = scan_path

    def scan_directory(self):
        print(f"Scanning directory: {self.scan_path}")
        malware_files = []
        for root, dirs, files in os.walk(self.scan_path):
            for file in files:
                file_path = os.path.join(root, file)
                if self.is_suspicious(file_path):
                    malware_files.append(file_path)
        return malware_files

    def is_suspicious(self, file_path):
       
        suspicious_extensions = ['.exe', '.bat', '.vbs']
        if any(file_path.endswith(ext) for ext in suspicious_extensions):
            return True
       
        return False

    def report_results(self, malware_files):
        if malware_files:
            print("Suspicious files detected:")
            for file in malware_files:
                print(file)
        else:
            print("No suspicious files detected.")



def is_suspicious(self, file_path):
   
    suspicious_extensions = ['.exe', '.bat', '.vbs']
    if any(file_path.endswith(ext) for ext in suspicious_extensions):
        return True
    
    
    if os.path.getsize(file_path) > 1000000:  
        return True

  
    with open(file_path, 'r', errors='ignore') as f:
        content = f.read()
        if 'malicious' in content:
            return True
    
    return False

if __name__ == "__main__":
   
    guardian = WhisperingGuardian(scan_path=r"C:\Users\diyan\OneDrive\Desktop\Hangul BG")
    
   
    suspicious_files = guardian.scan_directory()
    
   
    guardian.report_results(suspicious_files)
```

```python
class Library:
    def __init__(self, name, location):  
        self.name = name
        self.location = location
        self.manuscripts = []
    
    def add_manuscript(self, manuscript):
        self.manuscripts.append(manuscript)
    
    def find_manuscript(self, title):
        for manuscript in self.manuscripts:
            if manuscript.title == title:
                return manuscript
        return None

class Manuscript:
    def __init__(self, title, description, symbols):  
        self.title = title
        self.description = description
        self.symbols = symbols

    def __str__(self):  
        return f"Title: {self.title}, Description: {self.description}"

class Archivist:
    def __init__(self, name): 
        self.name = name

    def guard(self, library):
        return f"{self.name} is guarding the {library.name}."

class Programmer:
    def __init__(self, name):  
        self.name = name
    
    def discover_manuscript(self, library, title):
        manuscript = library.find_manuscript(title)
        if manuscript:
            return f"{self.name} discovered the manuscript: {manuscript}"
        else:
            return f"{self.name} couldn't find the manuscript titled '{title}'."


vault_of_whispers = Library("Vault of Whispers", "Silicon Valley")
whispering_code = Manuscript("The Whispering Code", "An ancient manuscript bound in leather, etched with moving symbols.", "Symbols that move when glanced at from the corner of the eye.")
vault_of_whispers.add_manuscript(whispering_code)

the_archivist = Archivist("The Archivist")
aria = Programmer("Aria")


print(the_archivist.guard(vault_of_whispers))
print(f"One late evening, Aria stumbled upon the {vault_of_whispers.name}.")
print(aria.discover_manuscript(vault_of_whispers, "The Whispering Code"))
```

```python
class Netiverse:
    def __init__(self):  
        self.data_flow = "Normal"
        self.anomalies = []
    
    def detect_anomalies(self):
        self.anomalies = ["Malus detected in Sector 7", "Malus detected in Sector 12"]
        return self.anomalies

    def neutralize_threats(self):
        if self.anomalies:
            self.anomalies = []
            self.data_flow = "Restored"
            return "All threats neutralized. Balance and order restored."
        else:
            return "No threats detected."

class GuardianProtocol:
    def __init__(self): 
        self.active = False
    
    def activate(self, netiverse):
        self.active = True
        anomalies = netiverse.detect_anomalies()
        return anomalies

class Guardian:
    def __init__(self, name):  
        self.name = name
        self.protocol = GuardianProtocol()

    def merge_with_system(self, netiverse):
        if self.protocol.activate(netiverse):
            return f"{self.name} has become one with the system, detecting and neutralizing threats."
        else:
            return f"{self.name} has become one with the system, but no threats were found."

class ManuscriptReader:
    def __init__(self, name):  
        self.name = name
    
    def read_manuscript(self, manuscript):
        return f"{self.name} read the manuscript: {manuscript}"


netiverse = Netiverse()
elara = Guardian("Elara")  
manuscript = ("It was a daunting decision, but Elara knew the stakes. The future of the Netiverse depended on her. "
              "With a heavy heart and steely determination, Elara initiated the process. The code enveloped her, and she felt her consciousness merging with the Guardian Protocol. "
              "It was a strange sensation, both exhilarating and terrifying. As she became one with the system, her senses expanded, and she could feel the ebb and flow of data throughout the Netiverse. "
              "The Guardian Protocol activated, and immediately, it began to detect anomalies. The Malus, hidden within the shadows, were revealed. "
              "Elara, now the Guardian, guided the system to neutralize the threats, restoring balance and order to the Netiverse.")
aria = ManuscriptReader("Aria")


print(elara.merge_with_system(netiverse))
print(netiverse.neutralize_threats())
print(aria.read_manuscript(manuscript))
print("As Aria read the final words of the manuscript, she felt a chill run down her spine.")
print("The tale was a cryptic reflection of her own world, where cyber threats lurked in the shadows, and the need for a robust malware detection system was more pressing than ever.")
```

```python
import os
import tkinter as tk
from tkinter import filedialog, messagebox
import hashlib
import json


MALWARE_SIGNATURES = {
    "d41d8cd98f00b204e9800998ecf8427e": "Example Malware 1",
    "e99a18c428cb38d5f260853678922e03": "Example Malware 2"
}

def compute_md5(file_path):
    """Compute MD5 hash of the given file."""
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

def scan_directory(directory):
    """Scan all files in the given directory for malware."""
    suspicious_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_hash = compute_md5(file_path)
            if file_hash in MALWARE_SIGNATURES:
                suspicious_files.append({
                    "path": file_path,
                    "hash": file_hash,
                    "name": MALWARE_SIGNATURES[file_hash]
                })
    return suspicious_files

def report_suspicious_files(suspicious_files):
    """Export suspicious files report to a JSON file."""
    if not suspicious_files:
        messagebox.showinfo("No Malware Detected", "No suspicious files were found.")
        return
    
    report_file = filedialog.asksaveasfilename(defaultextension=".json",
                                               filetypes=[("JSON Files", "*.json")],
                                               title="Save Report As")
    if not report_file:
        return

    with open(report_file, "w") as f:
        json.dump(suspicious_files, f, indent=4)
    
    messagebox.showinfo("Report Saved", f"Suspicious files report saved to {report_file}.")

def scan_and_report():
    """Function to handle the scanning and reporting process."""
    directory = filedialog.askdirectory(title="Select Directory to Scan")
    if not directory:
        return

    suspicious_files = scan_directory(directory)
    report_suspicious_files(suspicious_files)


root = tk.Tk()
root.title("Malware Detection System")
root.geometry("300x150")

def start_scan():
    """Start the scanning process."""
    scan_and_report()

scan_button = tk.Button(root, text="Scan Directory", command=start_scan)
scan_button.pack(pady=20)

root.mainloop()
```
